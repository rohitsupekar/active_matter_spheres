```python
import glob
import sys
sys.path.append('../')
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
import cartopy.crs as ccrs
from dedalus.extras import plot_tools
import logging
from matplotlib.animation import FFMpegWriter
logger = logging.getLogger(__name__)
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Parameters
first_frame = 1
figsize = (6, 2)
dpi = 300
show_time = True
gridlines = True
coastlines = False
edgecolor = 'k'
proj = ccrs.PlateCarree(central_longitude=0)
proj = ccrs.Mollweide(central_longitude=0)
proj = ccrs.Orthographic(central_longitude=0, central_latitude=30)
fields = ['v_ph', 'om']
clim_mod = {'v_ph': 4, 'om': 50}

STRNAME1 = "sphere103"
STRNAME2 = "sphere104"
f_inertial = 300.

input_folder1 = "../data/%s" %(STRNAME1)
input_folder2 = "../data/%s" %(STRNAME2)
output_folder = "../videos"
Omega = f_inertial/2.

#count files in the input folder
last_frame = len(glob.glob1("".join([input_folder1,'/']),"*.npz"))
#last_frame = 50
# Setup output folder
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
UseTex = False

#set clims for all the fields
max_vals = {key: 0 for key in fields}
clims = {key: 0 for key in fields}
for field in fields:
    for i in range(first_frame, last_frame + 1, 1):
        with np.load("".join([input_folder1, '/output_%i.npz' %i])) as file:
            fieldval = file[field]
            max_vals[field] = max(max_vals[field], np.max(fieldval))

for field in fields:
    clims[field] = 0.75*max_vals[field]

metadata = dict(title='Movie', artist='Matplotlib', comment='Movie support!')
writer = FFMpegWriter(fps=15, metadata=metadata)

for field in fields:
    
    # Setup figure projection
    fig, axs = plt.subplots(1,3, figsize=figsize, subplot_kw={'projection': proj})
    plt.rc('text', usetex=False)

    with writer.saving(fig, "%s/%s_%s_comp.mp4" %(output_folder, STRNAME1, field), dpi):
        for i in range(first_frame+1, last_frame + 1, 1):

            # Load data
            print('Frame:%i'%(i))
            for axes in axs:
                axes.cla()
            
            ####### Inertial simulation ########
            with np.load("".join([input_folder1, '/output_%i.npz' %i])) as file:
                phi = file['phi']
                theta = file['theta']
                time = file['t'][0]
                thth, phiphi = np.meshgrid(theta, phi)

                data = file[field]
                signs = [0, -1]
                count = 0
                for axes in axs[0:2]:
                    sign = signs[count]
                    # Create plot
                    lon = (phi + phi[1]/2 - np.pi + sign*Omega*time) * 180 / np.pi
                    lat = (np.pi/2 - theta) * 180 / np.pi
                    xmesh, ymesh = plot_tools.quad_mesh(lon, lat)
                    if field=='v_ph':
                        data_mod = data + sign*Omega*np.sin(thth)
                    elif field=='om':
                        data_mod = data - 2*sign*Omega*np.cos(thth)
                    else:
                        data_mod = data
                        
                    image = axes.pcolormesh(xmesh, ymesh, data_mod.T, cmap='RdBu_r', transform=ccrs.PlateCarree())
                   
                    if gridlines:
                        axes.gridlines(xlocs=np.arange(0, 361, 30), ylocs=np.arange(-60, 61, 30), color='k')

                    axes.set_global()
                    axes.outline_patch.set_edgecolor(edgecolor)
                    if count==0: axes.set_title('Inertial', fontsize=10)
                    else: axes.set_title('Rotating Frame', fontsize=10)
                        
                    count += 1
                    
                    if clims:
                        image.set_clim(-clims[field], clims[field])
                        if sign!=0: image.set_clim(-clim_mod[field], clim_mod[field])
                    else:
                        clim_i = np.max(np.abs(data))
                        image.set_clim(-clim_i, clim_i)
                        
                    if i == first_frame+1:
                        fig.colorbar(image, ax=axes)
                    
                        
            ####### Rotating simulation ########
            with np.load("".join([input_folder2, '/output_%i.npz' %i])) as file:
                phi = file['phi']
                theta = file['theta']
                time = file['t'][0]
                thth, phiphi = np.meshgrid(theta, phi)

                data = file[field]
                count = 0
                for axes in axs[2:]:
                    sign = signs[count]; count += 1
                    # Create plot
                    lon = (phi + phi[1]/2 - np.pi) * 180 / np.pi
                    lat = (np.pi/2 - theta) * 180 / np.pi
                    xmesh, ymesh = plot_tools.quad_mesh(lon, lat)
              
                    image = axes.pcolormesh(xmesh, ymesh, data.T, cmap='RdBu_r', transform=ccrs.PlateCarree())
                    if show_time:
                        if field=='v_ph': title = fig.suptitle(r'$v_\phi (t = %.4f, f = %.1f)$' %(time, f_inertial), usetex=UseTex)
                        if field=='om': title = fig.suptitle(r'$\omega (t = %.4f, f = %.1f)$' %(time, f_inertial), usetex=UseTex)
                    if gridlines:
                        axes.gridlines(xlocs=np.arange(0, 361, 30), ylocs=np.arange(-60, 61, 30), color='k')

                    axes.set_global()
                    axes.outline_patch.set_edgecolor(edgecolor)
          
                    if clims:
                        image.set_clim(-clim_mod[field], clim_mod[field])
                    else:
                        clim_i = np.max(np.abs(data))
                        image.set_clim(-clim_i, clim_i)
                        
                    if i == first_frame+1:
                        fig.colorbar(image, ax=axes)
                        
                    axes.set_title('Rotating Sim', fontsize=10)
                        

                        
                del data, data_mod
                

            writer.grab_frame()
```
